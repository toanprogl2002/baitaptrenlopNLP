```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sentence classification - Sheet1.csv", encoding="utf8")
df
```

```python
from underthesea import word_tokenize
X1 = df['normal sentence'].dropna().astype(str)
x_nor = [word_tokenize(x, format = 'text').lower() for x in X1]
```

```python
y_nor = []
for i in range(len(x_nor)):
    y_nor.append(0)
len(y_nor)
```

```python

X2 = df['aggressive sentence'].dropna().astype(str)
x_agg = [word_tokenize(x, format = 'text').lower() for x in X2]
```

```python
len(x_agg)
```

```python
for i in range(len(x_agg)):
    y_nor.append(1)
len(y_nor)
```

```python
new = pd.DataFrame()
new['x'] = df['normal sentence'].append(df['aggressive sentence']).reset_index(drop=True)
new['y'] = y_nor
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(new['x'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, new['y'], random_state=42,test_size=0.2)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Xây dựng mô hình SVM
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
print('SVM accuracy:', svm_acc)

# Xây dựng mô hình NB
nb = MultinomialNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
nb_acc = accuracy_score(y_test, nb_pred)
print('Naive Bayes accuracy:', nb_acc)

# Xây dựng mô hình kNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
print('k-Nearest Neighbors accuracy:', knn_acc)

# Xây dựng mô hình Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)
print('Logistic Regression accuracy:', lr_acc)
```

```python
# from nltk.corpus import stopwords
from collections import defaultdict
# nltk.download('stopwords')
import string
import nltk


# stop_words = set(stopwords.words('vietnamese'))
word_counts = defaultdict(int)
for doc in X_cau2:
    words = doc.translate(str.maketrans('', '', string.punctuation)).lower().split()
    words = ['<s>'] + words
    for i in range(1, len(words)):
        prev_word = words[i-1]
        curr_word = words[i]
        word_counts[(prev_word, curr_word)] += 1
```

```python
word = "<s> hôm nay trời nắng."
words = word.split()
words
# word.split
# for i in word:
    # print(i)
```

```python
vocab_size = len(set(words))
probs = np.zeros((vocab_size, vocab_size))
for i, prev_word in enumerate(set(words)):
    for j, curr_word in enumerate(set(words)):
        count = word_counts.get((prev_word, curr_word), 0)
        probs[i][j] = (count + 1) / (sum([count for (w1, w2), count in word_counts.items() if w1 == prev_word]) + vocab_size)
```

```python
for i, prev_word in enumerate(set(words)):
    for j, curr_word in enumerate(set(words)):
        print(f'P({prev_word} , {curr_word}) = {probs[i][j]:.4f}')
```
