```python
with open('10M.txt', encoding="utf8") as f:
    lines = f.readlines()
```

```python
print(len(lines))
linesNew = []
for i in range(100):
    linesNew.append(lines[i])
linesNew
```

```python
from underthesea import word_tokenize
from collections import Counter

#  = []
lenText = []
for line in linesNew:
    word = word_tokenize(line.lower(), format="text")
    # line.append(word)
    # print(len(word))
    lenText.append(len(word))

# words
```

```python
import matplotlib.pyplot as plt
plt.hist(lenText, edgecolor='black')
plt.xlabel('weight')
plt.ylabel('Person count')
plt.show()
```

```python
count = dict()
for line in linesNew:
    word = word_tokenize(line.lower())
    for w in word:
        if w in count:
            count[w] += 1
        else:
            count[w] = 1
count
```

```python
from nltk.probability import FreqDist
count = FreqDist(count)
count.plot(50, cumulative=True)
```

```python
from itertools import islice

sort_orders = sorted(count.items(), key=lambda x: x[1], reverse=True)

def take(n, iterable):
    """Return the first n items of the iterable as a list."""
    return list(islice(iterable, n))

take(10,count)
```
